```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
#1. Load and preprocess the image
image1 = cv2.imread('LEENA.jpg', cv2.IMREAD_GRAYSCALE) # Load grayscale
image =cv2.resize(image1, (28, 28)) # Resize to 28x28
image= image / 255.0 # Normalize pixel values to [0,1]
x = image.flatten() # Flatten to vector (784,)
#2. Display the input image
plt.imshow(image1, cmap="gray")
plt.title("Input Image (Original)")
plt.axis('off')
plt.show()
plt.imshow(image, cmap="gray")
plt.title("Input Image (28x28 Normalized)")
plt.axis('off')
plt.show()
print("Input Vector Shape:", x.shape)
print("Input Vector (first 20 values):", np.round(x[:20], 2))
#3. Initialize a simple FNN (784 input 64 hidden  10 output)
np.random.seed (42)
W1 = np.random.randn(784, 64) 
b1 = np.zeros(64)
W2= np.random.randn(64, 10) 
b2 = np.zeros (10)
#4. Define activation functions
def relu(z):
    return np.maximum(0, z)
def step(z):
    return np.where (z > 0, 1, 0)


#5. Forward Pass
#Layer 1: Input Hidden
z1 = np.dot(x, W1) + b1
a1  = relu(z1)
# Display hidden layer output
print("\nHidden Layer Output (first 10 values):", np.round(a1[:10], 3))
#Layer 2: Hidden Output
z2  = np.dot(a1, W2) + b2
output  = step (z2) # Step function applied here
#6. Display final prediction
print("\nOutput (Binary Activations):", output)
print("Predicted Class(es):", np.where (output == 1) [0]) # may return multiple if more than one is activated
```
