## ## An end-to-end example
Let's build a convolutional neural network to find patterns in our images, more specifically we a need way to:

* Load our images
* preprocess our images
* Build a CNN to find patterns in our images
* Compile our CNN
* fit the cnn to our training data

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
tf.random.set_seed(42)

train_datagen = ImageDataGenerator(rescale=1./255)
valid_datagen = ImageDataGenerator(rescale=1./255)

train_dir = "C:/Users/sakth/OneDrive/Desktop/ml_saga/dataset/pizza_steak/train"
test_dir = "C:/Users/sakth/OneDrive/Desktop/ml_saga/dataset/pizza_steak/test"

train_data = train_datagen.flow_from_directory(train_dir,
                                               batch_size = 32,
                                               target_size = (244,244),
                                               class_mode = "binary",
                                               seed = 42
                                               )

valid_data = valid_datagen.flow_from_directory(test_dir,
                                                  batch_size=32,
                                                  target_size = (244,244),
                                                  class_mode = "binary",
                                                  seed = 42
                                                  )



model_1 = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(filters=10,
                           kernel_size = 3,
                           activation = 'relu',
                           input_shape = (244,244,3)),
    tf.keras.layers.Conv2D(10,3,activation="relu"),
    tf.keras.layers.MaxPool2D(padding="valid",pool_size=2),
    tf.keras.layers.Conv2D(10,3,activation="relu"),
    tf.keras.layers.Conv2D(10,3,activation="relu"),
    tf.keras.layers.MaxPool2D(2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(1,activation="sigmoid")
])

model_1.compile(loss = "binary_crossentropy",optimizer = tf.keras.optimizers.Adam(),metrics =["accuracy"])

with tf.device("/GPU:0"):
    history_1 = model_1.fit(
        train_data,
        epochs=5,
        steps_per_epoch=len(train_data),
        validation_data=valid_data,
        validation_steps=len(valid_data)
    )

## 1500/32 = 47 means each epoch will run 47 iteration 32 images
```

```python
def create_model(learning_rate=0.001, filters=10, dense_units=1):
    model = tf.keras.Sequential([
        tf.keras.layers.Conv2D(filters, 3, activation='relu', input_shape=(244,244,3)),
        tf.keras.layers.Conv2D(filters, 3, activation='relu'),
        tf.keras.layers.MaxPool2D(2),
        tf.keras.layers.Conv2D(filters*2, 3, activation='relu'),
        tf.keras.layers.Conv2D(filters*2, 3, activation='relu'),
        tf.keras.layers.MaxPool2D(2),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(dense_units, activation='relu'),
        tf.keras.layers.Dense(1, activation='sigmoid')
    ])
    
    model.compile(loss='binary_crossentropy',
                 optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
                 metrics=['accuracy'])
    return model

# Test different hyperparameters
hyperparams = [
    {'learning_rate': 0.001, 'filters': 16, 'dense_units': 32},
    {'learning_rate': 0.0001, 'filters': 32, 'dense_units': 64},
    {'learning_rate': 0.01, 'filters': 8, 'dense_units': 16}
]

for i, params in enumerate(hyperparams):
    print(f"Training model with params: {params}")
    model = create_model(**params)
    
    history = model.fit(
        train_data,
        epochs=3,  # Fewer epochs for testing
        steps_per_epoch=len(train_data),
        validation_data=valid_data,
        validation_steps=len(valid_data),
        verbose=0
    )
    
    val_acc = history.history['val_accuracy'][-1]
    print(f"Model {i+1} - Validation Accuracy: {val_acc:.4f}\n")
```
