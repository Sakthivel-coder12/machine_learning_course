## ## An end-to-end example
Let's build a convolutional neural network to find patterns in our images, more specifically we a need way to:

* Load our images
* preprocess our images
* Build a CNN to find patterns in our images
* Compile our CNN
* fit the cnn to our training data

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
tf.random.set_seed(42)

train_datagen = ImageDataGenerator(rescale=1./255)
valid_datagen = ImageDataGenerator(rescale=1./255)

train_dir = "C:/Users/sakth/OneDrive/Desktop/ml_saga/dataset/pizza_steak/train"
test_dir = "C:/Users/sakth/OneDrive/Desktop/ml_saga/dataset/pizza_steak/test"

train_data = train_datagen.flow_from_directory(train_dir,
                                               batch_size = 32,
                                               target_size = (224,224),
                                               class_mode = "binary",
                                               seed = 42
                                               )

valid_data = valid_datagen.flow_from_directory(test_dir,
                                                  batch_size=32,
                                                  target_size = (224,224),
                                                  class_mode = "binary",
                                                  seed = 42
                                                  )



model_1 = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(filters=10,
                           kernel_size = 3,
                           activation = 'relu',
                           input_shape = (224,224,3)),
    tf.keras.layers.Conv2D(10,3,activation="relu"),
    tf.keras.layers.MaxPool2D(padding="valid",pool_size=2),
    tf.keras.layers.Conv2D(10,3,activation="relu"),
    tf.keras.layers.Conv2D(10,3,activation="relu"),
    tf.keras.layers.MaxPool2D(2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(1,activation="sigmoid")
])

model_1.compile(loss = "binary_crossentropy",optimizer = tf.keras.optimizers.Adam(),metrics =["accuracy"])

with tf.device("/GPU:0"):
    history_1 = model_1.fit(
        train_data,
        epochs=5,
        steps_per_epoch=len(train_data),
        validation_data=valid_data,
        validation_steps=len(valid_data)
    )

## 1500/32 = 47 means each epoch will run 47 iteration 32 images
```

## Hyper_Parameter tunning

```python
def create_model(learning_rate=0.001, filters=10, dense_units=1):
    model = tf.keras.Sequential([
        tf.keras.layers.Conv2D(filters, 3, activation='relu', input_shape=(224,224,3)),
        tf.keras.layers.Conv2D(filters, 3, activation='relu'),
        tf.keras.layers.MaxPool2D(2),
        tf.keras.layers.Conv2D(filters*2, 3, activation='relu'),
        tf.keras.layers.Conv2D(filters*2, 3, activation='relu'),
        tf.keras.layers.MaxPool2D(2),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(dense_units, activation='relu'),
        tf.keras.layers.Dense(1, activation='sigmoid')
    ])
    
    model.compile(loss='binary_crossentropy',
                 optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
                 metrics=['accuracy'])
    return model

# Test different hyperparameters
hyperparams = [
    {'learning_rate': 0.001, 'filters': 16, 'dense_units': 32},
    {'learning_rate': 0.0001, 'filters': 32, 'dense_units': 64},
    {'learning_rate': 0.01, 'filters': 8, 'dense_units': 16}
]

for i, params in enumerate(hyperparams):
    print(f"Training model with params: {params}")
    model = create_model(**params)
    
    history = model.fit(
        train_data,
        epochs=3,  # Fewer epochs for testing
        steps_per_epoch=len(train_data),
        validation_data=valid_data,
        validation_steps=len(valid_data),
        verbose=0
    )
    
    val_acc = history.history['val_accuracy'][-1]
    print(f"Model {i+1} - Validation Accuracy: {val_acc:.4f}\n")
```

```python
## plot the validation and trining curves separately
import matplotlib.pyplot as plt
def plot_loss_curves(history):
    """returns separate loss curves for training and validtio metrics"""
    loss = history.history["loss"]
    val_loss = history.history["val_loss"]
    accuracy = history.history["accuracy"]
    val_accuracy = history.history["val_accuracy"]
    epochs = range(len(history.history["loss"])) # how many epochs did we run for..?
    
    
    ## plot the loss
   
    plt.plot(epochs, loss,label = "training_loss")
    plt.plot(epochs,val_loss,label ="val_loss")
    plt.title("loss")
    plt.xlabel("epochs")
    plt.legend()
    
    # plot the accuracy
    plt.figure()
    plt.plot(epochs,accuracy,label = "training_loss")
    plt.plot(epochs,val_accuracy,label = "val_accuracy")
    plt.title('accuracy')
    plt.xlabel("epochs")
    plt.legend()
```

> Note : When the model's validation loss starts to increases , it's likely that the is overfitting the training dataset. This means, it's learning the patterns in the traning dataset too well and thus the model's ability to generalize to unseen data will be diminished

```python
plot_loss_curves(history_4)
```

```python
model_4.evaluate(test_data)
```

Note : Ideally the two loss curves(training and validation) will be very similar to eachother decreasing at similar ratess, where there are large differences you model may be overfitting.

## 6. Adjust the model parameters

Fitting a machine learning model comes in 3 steps:

0. create a baseline 
1. Beat the baseline by overfitting a larger model
2. Reduce overifitting


**Ways to induce**
* Increase the number of conv layers
* Increse the number of conv filters
* Add another dense layer to the output of our flattened layer

Redcue Overfitting:
* Add data augmentation
* Add regularization layers(such as MaxPool2D)
* Add more data..


Reducting the overfitting is also know as *Regularization*

## Opening our bag of tricks and finding data augmentation 


```python
## Create the ImageDatatGenerator training instance with data augmentation

train_datagen_augmented = ImageDataGenerator(rescale=1/255.,
                                     rotation_range = 0.2, # how much do you want to rotate an image ?
                                     shear_range = 0.2, # how much do yo want to shear an image ?
                                     zoom_range = 0.2, # zoom in randomly on an image
                                     width_shift_range = 0.2, #move your image around on the x-axis
                                     height_shift_range = 0.3, # move your image around on the y-axis
                                     horizontal_flip = True # do you want to flip and  an image?
                                     )
## Create the ImageDataGenerator testing instance with data augmentation
train_datagen_2 = ImageDataGenerator(rescale = 1/255.)
test_datagen_2 = ImageDataGenerator(rescale=1/255.) 
```

> **What is data augmentation..?**
Data augmentation is the process of altering our training data, leading it to have more diversity and in turn allowing our model to learn more generalizable(hopefully) patterns. Altering might mean adjusting the rotation of an image, flipping it , cropping it or something similar.

Let's write some code the visualize the data augmentation...

```python
## Import data and augment it from training directory
print("Augmented traning data")
train_data_augmented = train_datagen_augmented.flow_from_directory(
    directory = train_dir,
    target_size = (224,224),
    class_mode = "binary",
    batch_size = 32,
    shuffle = False
)
print("Non-augmented training data")
train_data = train_datagen_2.flow_from_directory(
    train_dir,
    target_size = (224,224),
    class_mode = "binary",
    batch_size = 32,
    shuffle = False
)

print("Non-augmented test data: ")

test_data = test_datagen_2.flow_from_directory(
    test_dir,
    target_size = (224,224),
    batch_size = 32,
    class_mode = "binary"
)
```

> **Note:** Data augmentation is usually only performed on the training data. Using `ImageDataGenerator` build-in data augmentation parameters our images are left as they are in the directories but are modified as they're loaded into the model.
