```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
%matplotlib inline
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
x = df[['tip','sex','smoker','day','time','size']]
y = df['total_bill']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
x_train.info()
```

```python
x_train['sex'].value_counts(),x_train['day'].value_counts()
```

```python
x_train['time'].value_counts()
```

```python
## We have to change the catetory into integers 
from sklearn.preprocessing import LabelEncoder
lst = ['sex','smoker','time']
for i in lst:
    le = LabelEncoder()
    x_train[i] = le.fit_transform(x_train[i])
    x_test[i] = le.transform(x_test[i])
```

```python
x_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x_train['sex']= le.fit_transform(x_train['sex'])
x_test['sex'] = le.transform(x_test['sex'])
x_train['time'] = le.fit_transform(x_train['time'])
x_test['time'] = le.transform(x_test['time'])
x_train['smoker'] = le.fit_transform(x_train['smoker'])
x_test['smoker'] = le.transform(x_test['smoker'])
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
transform = ColumnTransformer(transformers=[('onehot',OneHotEncoder(drop='first'),[3])],remainder='passthrough') ## 3 is index of day
```

```python
import sys
np.set_printoptions(threshold=sys.maxsize)
x_train = transform.fit_transform(x_train)
```

```python
x_test = transform.transform(x_test)
```

```python
x_test
```

```python
from sklearn.svm import SVR
from sklearn.metrics import r2_score,mean_absolute_error
svr = SVR()
svr.fit(x_train,y_train)
y_pred = svr.predict(x_test)
```

```python
print(r2_score(y_test,y_pred))

print(mean_absolute_error(y_test,y_pred))
```

```python
## Hypterparameter tuning using the Gridsearchcv
from sklearn.model_selection import GridSearchCV
param_grid = {
    'C' : [0.1,100,1,10,1000],
    'gamma' : [1,0.1,0.01,0.001,0.0001],
    'kernel' :['rbf']
}
```

```python
grid = GridSearchCV(SVR(),param_grid=param_grid)
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid_pred = grid.predict(x_test)
```

```python
print(r2_score(y_test,grid_pred))

print(mean_absolute_error(y_test,grid_pred))
```
