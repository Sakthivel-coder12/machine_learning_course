```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
df = load_iris()
```

```python
df
```

```python
print(df['DESCR'])
```

```python
print(df['target'])
```

```python
x = pd.DataFrame(df['data'],columns=['spl','spw','pel','pew'])
```

```python
x
```

```python
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
x_train,x_test,y_train,y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier.fit(x_train,y_train)
```

```python
## visualize the decision tree
import warnings

from sklearn.tree import plot_tree
plt.figure(figsize=(10,20))
plot_tree(treeclassifier,filled=True)
warnings.filterwarnings("ignore")
```

```python
y_pred = treeclassifier.predict(x_test)
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
path = treeclassifier.cost_complexity_pruning_path(x_train, y_train)
path = treeclassifier.cost_complexity_pruning_path(x_train, y_train)
ccp_alphas = path.ccp_alphas
print(ccp_alphas)
# Train multiple trees and plot
import matplotlib.pyplot as plt

acc = []
for alpha in ccp_alphas:
    model = DecisionTreeClassifier(random_state=0, ccp_alpha=alpha)
    model.fit(x_train, y_train)
    acc.append(model.score(x_test, y_test))

plt.plot(ccp_alphas, acc, marker='o')
plt.xlabel('ccp_alpha')
plt.ylabel('Accuracy')
plt.title('Post-Pruning Performance')
plt.show()
```

```python
path.ccp_alphas
```

```python
## The hyperparameter post pruning

from sklearn.tree import DecisionTreeClassifier
treeclassifier = DecisionTreeClassifier(max_depth=2)

treeclassifier.fit(x_train,y_train)

import warnings

from sklearn.tree import plot_tree
plt.figure(figsize=(10,20))
plot_tree(treeclassifier,filled=True)
warnings.filterwarnings("ignore")
```

```python
y_pred = treeclassifier.predict(x_test)
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

## By using the pre-pruning

```python
param = {
    'criterion' : ['gini','entropy','log_loss'],
    'splitter' : ['best','random'],
    'max_depth' : [1,2,3,4,5],
    'max_features' : ['auto','sqrt','log2']
}
clf = DecisionTreeClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(clf,param_grid=param,cv=5,scoring='accuracy')
grid.fit(x_train,y_train)
y_p = grid.predict(x_test)
```

```python
y_p
```

```python
print(accuracy_score(y_test,y_p))
print(classification_report(y_test,y_p))
print(confusion_matrix(y_test,y_p))
```
