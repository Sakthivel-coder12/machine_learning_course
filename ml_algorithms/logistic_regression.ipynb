```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, StratifiedKFold

model = LogisticRegression(max_iter=5000)

# Define valid param combinations manually
param_grid = [
    {'penalty': ['l2'], 'C': [100, 10, 1.0, 0.1, 0.01], 'solver': ['sag', 'saga']},
    {'penalty': ['l1'], 'C': [100, 10, 1.0, 0.1, 0.01], 'solver': ['saga']},
    {'penalty': ['elasticnet'], 'C': [100, 10, 1.0, 0.1, 0.01], 'solver': ['saga'], 'l1_ratio': [0.5]}
]

cv = StratifiedKFold()

grid = GridSearchCV(estimator=model, param_grid=param_grid, scoring='accuracy', cv=cv, n_jobs=-1)
grid.fit(x_train, y_train)

# Print the best results
print("Best Score: ", grid.best_score_)
print("Best Params: ", grid.best_params_)
```

```python
y_pred = grid.predict(x_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

## Logistic Regression in the imbalanced dataset

```python
from collections import Counter
from sklearn.datasets import make_classification
```

```python
x,y = make_classification(n_samples=1000,n_classes=2,n_clusters_per_class=1,weights=[0.99],n_redundant=0,random_state=10)
```

```python
x
```

```python
y
```

```python
sns.scatterplot(x = pd.DataFrame(x)[0],y = pd.DataFrame(x)[1],hue = y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
## Hyperparamter tuning
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
penalty=['l1', 'l2', 'elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
class_weight=[{0:w,1:y} for w in [1,10,50,100] for y in [1,10,50,100]]
```

```python
params = dict(penalty = penalty,C = c_values, solver = solver ,class_weight = class_weight)
```

```python
params
```

```python
## Grid searchCv
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
grid = GridSearchCV(estimator = model,param_grid=params,scoring='accuracy',cv = cv,n_jobs=-1)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_estimator_
```

```python
grid. best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(x_test)
y_pred
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

## Logistic Regression with ROC curve and ROC AUC score

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score,roc_curve
```

```python
x,y = make_classification(n_samples=1000,n_classes=2,random_state=1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
dummy_model_prob = [0 for _ in range(len(y_test))]
dummy_model_prob
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
model_prob = model.predict_proba(x_test)
```

```python
model_prob
```

```python
model_prob = model_prob[:,1]
```

```python
dummy_model = roc_auc_score(y_test,dummy_model_prob)
model_auc = roc_auc_score(y_test,model_prob)
print(dummy_model)
print(model_auc)
```

```python
dummy_fpr, dummy_tpr, _ = roc_curve(y_test, dummy_model_prob)
model_fpr, model_tpr, thresholds = roc_curve(y_test, model_prob)
```

```python
thresholds
```

```python
# plot the roc curve for the model
plt.plot(dummy_fpr, dummy_tpr, linestyle='--', label='Dummy Model')
plt.plot(model_fpr, model_tpr, marker='.', label='Logistic')
# axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
# show the legend
plt.legend()
# show the plot
plt.show()
```

```python
# plot the roc curve for the model
import numpy as np
fig = plt.figure(figsize=(20,50))
plt.plot(dummy_fpr, dummy_tpr, linestyle='--', label='Dummy Model')
plt.plot(model_fpr, model_tpr, marker='.', label='Logistic')
ax = fig.add_subplot(111)
for xyz in zip(model_fpr, model_tpr,thresholds):   
    ax.annotate('%s' % np.round(xyz[2],2), xy=(xyz[0],xyz[1]))
# axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
# show the legend
plt.legend()
# show the plot
plt.show()
```
