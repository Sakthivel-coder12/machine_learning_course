```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
print(diabetes['DESCR'])
```

```python
diabetes['target']
```

```python
df = pd.DataFrame(diabetes.data,columns=['age',
    'sex',
    'bmi',
    'bp',
    's1',
    's2',
    's3',
    's4',
    's5',
    's6'])
```

```python
df
```

```python
x = df
y = diabetes['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
x_train.corr()
```

```python
sns.heatmap(x_train.corr(),annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
clfr = DecisionTreeRegressor()
clfr.fit(x_train,y_train)
y_pred = clfr.predict(x_test)
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
import warnings

from sklearn.tree import plot_tree
plot_tree(clfr,filled=True)
warnings.filterwarnings('ignore')
```

```python
para = {
    'criterion' : ['squared_error','friend_mse','absolute_error'],
    'splitter' : ['best','random'],
    'max_depth' : [1,2,3,4,5,10,10,25],
    'max_features' : ['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(clfr,param_grid=para,cv = 5,scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(x_train,y_train)
```

```python
grid_pred = grid.predict(x_test)
```

```python
print(r2_score(y_test,grid_pred))
print(mean_absolute_error(y_test,grid_pred))
print(mean_squared_error(y_test,grid_pred))
```

```python
grid.best_estimator_
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
selectedregressor = DecisionTreeRegressor(criterion='squared_error',max_depth=5,max_features='sqrt',splitter='best')
selectedregressor
```

```python
selectedregressor.fit(x_train,y_train)
sepred = selectedregressor.predict(x_test)
```

```python
print(r2_score(y_test,sepred))
print(mean_absolute_error(y_test,sepred))
print(mean_squared_error(y_test,sepred))
```

```python
plt.figure(figsize=(20,20))
plot_tree(selectedregressor,filled=True)
warnings.filterwarnings('ignore')
```
